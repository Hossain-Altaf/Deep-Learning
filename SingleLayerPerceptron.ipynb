```python

# Single Neuron with Forward & Backward Propagation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score


# Load Dataset
df = pd.read_csv("cancer.csv")
print(df.head())

# Separate features and target
X = df.iloc[:, :-1]   #Selects all columns except the last
y = df.iloc[:, -1]    #Selects last column

# Convert target labels to 0 and 1
y = pd.factorize(y)[0]

# One-hot encode categorical features
X = pd.get_dummies(X, drop_first=True)    #Converts categorical text into binary columns

X = X.values

# Split Dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    test_size=0.2,     #20% testing  , 80% training
    random_state=42    #Keeps split consistent every run
)

# Normalize Features
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)    #x′=(x−μ)/σ  	​
X_test = scaler.transform(X_test)          #Uses same mean and standard deviation from training data


# Neural Neuron
class SingleNeuron:

    def __init__(self, learning_rate=0.01, epochs=500):

        self.lr = learning_rate
        self.epochs = epochs       #number of training cycles.


    # Activation Function
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))     #coverts any number into 0(pos) or 1(neg)

    # Sigmoid Derivative
    def sigmoid_derivative(self, a):
        return a * (1 - a)              #Needed for backpropagation


    # Training
    def fit(self, X, y):

        n_samples, n_features = X.shape   #row colm

        self.weights = np.random.randn(n_features)    #Creates random weights.
        self.bias = np.random.randn()

        self.loss_history = []

        for epoch in range(self.epochs):

            total_loss = 0

            for i in range(n_samples):    #Processes one training sample at a time.


                # FORWARD PROPAGATION
                z = np.dot(X[i], self.weights) + self.bias    #z=w1​x1​+w2​x2​+⋯+b
                a = self.sigmoid(z)    #a=σ(z)

                # Binary Cross Entropy
                loss = -(
                    y[i] * np.log(a + 1e-10)              #L=−(yloga+(1−y)log(1−a))
                    + (1 - y[i]) * np.log(1 - a + 1e-10)
                )
                total_loss += loss      #Adds current sample loss


                # BACKWARD PROPAGATION

                delta = a - y[i]      #δ=a−y

                grad_w = delta * X[i]      #(∂w/∂L)=δx
                grad_b = delta


                # GRADIENT DESCENT
                self.weights -= self.lr * grad_w    #w=w−η(∂w/∂L​)
                self.bias -= self.lr * grad_b       #b=b−η(∂b/∂L)​


            self.loss_history.append(total_loss / n_samples)  #Stores average loss.

            if epoch % 25 == 0:     #prints every 50 epoch
                print(
                    f"Epoch {epoch:3d} "
                    f"Loss = {self.loss_history[-1]:.4f}"
                )


    # Prediction

    def predict(self, X):

        z = np.dot(X, self.weights) + self.bias   #Forward propagation abar
        a = self.sigmoid(z)                       #Probability output

        return (a >= 0.5).astype(int)    #Decision rule: >= 0.5 → 1 && < 0.5  → 0


# Train Model

model = SingleNeuron(
    learning_rate=0.05,
    epochs=500
)

model.fit(X_train, y_train)    #Starts training.


# Predictions

predictions = model.predict(X_test)     #Predicts unseen data

# Accuracy

accuracy = accuracy_score(y_test, predictions)   #Accuracy=(Total Predictions / Correct Predictions)


print("\nRESULTS:")
print("Weights:", model.weights)
print("Bias:", model.bias)
print(f"Accuracy: {accuracy * 100:.2f}%")


# Confusion Matrix

cm = confusion_matrix(y_test, predictions)

print("\nConfusion Matrix:")
print(cm)


# Learning Curve

#plt.figure(figsize=(8,5))
#plt.plot(model.loss_history)
#plt.xlabel("Epoch")
#plt.ylabel("Loss")
#plt.title("Training Loss")
#plt.grid(True)
#plt.show()
```
